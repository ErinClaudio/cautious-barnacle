```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show, output_file
```

```python
start1= datetime.datetime(2018,10,30)
end1=datetime.datetime(2019,5,30)
df = data.DataReader("TSLA", "yahoo",start=start1,end=end1)
df
```

```python
def value(c,o):
    if c > o:
        value="Increase"
    elif c<o:
        value= "Decrease"
    else:
        value="equal"
    return value

df["Status"]=[value(c,o) for c,o in zip(df.Close,df.Open)]
```

```python
df["Middle"]= (df.Open+df.Close)/2
df["Height"]=abs(df.Close-df.Open)
```

```python
df
```

```python
p = figure(x_axis_type="datetime",width=1000,height=300)
p.title.text="Candlestick Chart"

hours_12=12*60*60*1000

p.rect(df.index[df.Status=="Increase"],df.Middle[df.Status =="Increase"],
       hours_12,df.Height[df.Status =="Increase"],fill_color="blue", line_color="black")

p.rect(df.index[df.Status=="Decrease"],df.Middle[df.Status =="Decrease"],
       hours_12,df.Height[df.Status =="Decrease"], fill_color="red", line_color="black")

output_file("CS.html")
show(p)
```
